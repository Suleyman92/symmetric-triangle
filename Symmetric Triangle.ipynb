```python
import tkinter as tk
from tkinter.simpledialog import askinteger, askstring

def sym_tri(n):
    triangle = []

    for i in range(n):
        spaces = " " * (n - i - 1)
        stars = "* " * (i + 1)
        row = spaces + stars.rstrip()
        triangle.append(row)

    return triangle

def col(color):
    try:
        tk.Tk().option_add("*TButton*background", color)
        return True
    except tk.TclError:
        return False

def get_color():
    while True:
        color = askstring("Renk seçimi", "Bir rengir renk girin:")
        if color is None:
            return None
        elif col(color):
            return color
        else:
            print("Yalnış bir renk seçimi yaptınız! Lütfen tekrar deneyin")

def tri_size(triangle):
    max_length = len(triangle[-1])
    canvas_height = len(triangle) * 20
    canvas_width = max_length * 20
    return canvas_height, canvas_width

n = askinteger("Üçgen", "Üçgen boyutunu girin lütfen:")

if n is not None:
    tri = sym_tri(n)

    star_color = get_color()

    if star_color is not None:
        root = tk.Tk()
        root.title("Üçgen")

        canvas_height, canvas_width = tri_size(tri)
        canvas = tk.Canvas(root, height=canvas_height, width=canvas_width, bg="black")
        canvas.pack()

        y_position = 10
        for row in tri:
            x_position = canvas_width // 2 - (len(row) * 10)
            for char in row:
                if char == '*':
                    canvas.create_text(x_position, y_position, text=char, font=("Courier", 12, "bold"), fill=star_color)
                    x_position += 20
            y_position += 20  
            x_position = canvas_width // 2 - (len(row) * 10)  

        root.mainloop()
```
